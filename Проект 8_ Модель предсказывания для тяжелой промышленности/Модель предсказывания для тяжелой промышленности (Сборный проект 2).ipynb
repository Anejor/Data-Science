### –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å—á–µ—Ç–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è

```python
#—Å–æ–∑–¥–∞—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è:
def recovery_calc(rougher_output, rougher_input, rougher_tail):
    recovery = (rougher_output * (rougher_input - rougher_tail)) \
    / (rougher_input * (rougher_output - rougher_tail)) * 100
    return recovery
```

```python
#–≤—ã—á–∏—Å–ª—è—é —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery:
recovery = recovery_calc(train['rougher.output.concentrate_au'],
                         train['rougher.input.feed_au'],
                         train['rougher.output.tail_au'])
recovery
```

```python
#–Ω–∞—Ö–æ–∂—É MAE –º–µ–∂–¥—É —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery:
mae = mean_absolute_error(train['rougher.output.recovery'], recovery)
print('MAE –º–µ–∂–¥—É —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery =', mae)
```

    2. —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–∞ final.output.recovery.

### –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE

```python
#—Å–æ–∑–¥–∞—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è sMAPE:
def sMAPE_calc(y_target, y_predict):
    smape = np.mean(abs(y_target - y_predict) / ((abs(y_target) + abs(y_predict)) / 2)) * 100
    return smape
```

```python
#—Å–æ–∑–¥–∞—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ sMAPE:
def final_sMAPE_calc(smape_rougher, smape_final):
    return (0.25*smape_rougher + 0.75*smape_final)
```

```python
#–≤–≤–æ–∂—É —Å–≤–æ—é –º–µ—Ç—Ä–∏–∫—É —á–µ—Ä–µ–∑ make_scorer, —á—Ç–æ–±—ã –ø–æ—Ç–æ–º –≤ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ —É–∫–∞–∑–∞—Ç—å –∏ –Ω–µ –¥–µ–ª–∞—Ç—å –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é —Ä—É–∫–∞–º–∏:
score = make_scorer(sMAPE_calc, greater_is_better = False)
```

```python
#–¥–ª—è –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –ø—Ä–æ—Ü–µ—Å—Å–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö –ø—Ä–∏–º–µ–Ω—è—é make_pipeline –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞, –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π –∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:
model_rfr = make_pipeline(StandardScaler(), RandomForestRegressor(random_state=12345))

model_dtr = make_pipeline(StandardScaler(), DecisionTreeRegressor(random_state=12345))

model_lr = make_pipeline(StandardScaler(), LinearRegression())
```

```python
#–∑–∞–¥–∞—é –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –º–æ–¥–µ–ª–µ–π:
parameters_rfr = {'randomforestregressor__n_estimators':range(10, 60, 10),
                  'randomforestregressor__max_depth':range(1,7)}

parameters_dtr = {'decisiontreeregressor__max_depth':range(1,7)}

parameters_lr = {}
```

```python
#–∞–≥—Ä–µ–≥–∞—Ç GridSearchCV –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –¥–ª—è –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏:
grid_rougher_rfr = GridSearchCV(model_rfr, param_grid=parameters_rfr, cv=3, scoring=score, verbose=10, n_jobs=-1)
grid_final_rfr = GridSearchCV(model_rfr, param_grid=parameters_rfr, cv=3, scoring=score, verbose=10, n_jobs=-1)

grid_rougher_dtr = GridSearchCV(model_dtr, param_grid=parameters_dtr, cv=3, scoring=score, verbose=10, n_jobs=-1)
grid_final_dtr = GridSearchCV(model_dtr, param_grid=parameters_dtr, cv=3, scoring=score, verbose=10, n_jobs=-1)

grid_rougher_lr = GridSearchCV(model_lr, param_grid=parameters_lr, scoring=score, verbose=10, n_jobs=-1)
grid_final_lr = GridSearchCV(model_lr, param_grid=parameters_lr, scoring=score, verbose=10, n_jobs=-1)
```

```python
#–æ–±—É—á–∞—é –º–æ–¥–µ–ª–∏:
grid_rougher_rfr.fit(features_train, target_train_rougher)
grid_final_rfr.fit(features_train, target_train_final)

grid_rougher_dtr.fit(features_train, target_train_rougher)
grid_final_dtr.fit(features_train, target_train_final)

grid_rougher_lr.fit(features_train, target_train_rougher)
grid_final_lr.fit(features_train, target_train_final)
```

```python
#–æ–ø—Ä–µ–¥–µ–ª—è—é –ª—É—á—à—É—é –∏–∑ –º–æ–¥–µ–ª–µ–π –¥–ª—è 2 —Ü–µ–ª–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:
best_score_rougher_rfr = grid_rougher_rfr.best_score_
best_score_final_rfr = grid_final_rfr.best_score_

best_score_rougher_dtr = grid_rougher_dtr.best_score_
best_score_final_dtr = grid_final_dtr.best_score_

best_score_rougher_lr = grid_rougher_lr.best_score_
best_score_final_lr = grid_final_lr.best_score_

print('–ø–æ –º–æ–¥–µ–ª–∏ RandomForestRegressor –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç rougher =', -best_score_rougher_rfr)
print('–ø–æ –º–æ–¥–µ–ª–∏ DecisionTreeRegressor –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç rougher =', -best_score_rougher_dtr)
print('–ø–æ –º–æ–¥–µ–ª–∏ LinearRegression –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç rougher =', -best_score_rougher_lr)
print()
print('–ø–æ –º–æ–¥–µ–ª–∏ RandomForestRegressor –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç final =', -best_score_final_rfr)
print('–ø–æ –º–æ–¥–µ–ª–∏ DecisionTreeRegressor –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç final =', -best_score_final_dtr)
print('–ø–æ –º–æ–¥–µ–ª–∏ LinearRegression –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç final =', -best_score_final_lr)
```

```python
#–≤—ã—á–∏—Å–ª—è—é final sMAPE –¥–ª—è –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏:
final_smape_rfr = final_sMAPE_calc(best_score_rougher_rfr, best_score_final_rfr)
print('–ø–æ –º–æ–¥–µ–ª–∏ RandomForestRegressor –∏—Ç–æ–≥–æ–≤–æ–µ sMAPE =', -final_smape_rfr)

final_smape_dtr = final_sMAPE_calc(best_score_rougher_dtr, best_score_final_dtr)
print('–ø–æ –º–æ–¥–µ–ª–∏ DecisionTreeRegressor –∏—Ç–æ–≥–æ–≤–æ–µ sMAPE =', -final_smape_dtr)

final_smape_lr = final_sMAPE_calc(best_score_rougher_lr, best_score_final_lr)
print('–ø–æ –º–æ–¥–µ–ª–∏ LinearRegression –∏—Ç–æ–≥–æ–≤–æ–µ sMAPE =', -final_smape_lr)
```

```python
#–Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ –≤—ã—á–∏—Å–ª—è—é final sMAPE –ø–æ RandomForestRegressor:
features_test = np.array(test.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1))
target_test_rouger = np.array(test['rougher.output.recovery'])
target_test_final = np.array(test['final.output.recovery'])

test_rogher = grid_rougher_rfr.predict(features_test)
test_final = grid_final_rfr.predict(features_test)

rouger_smape_test = sMAPE_calc(target_test_rouger, test_rogher)
final_smape_test = sMAPE_calc(target_test_final, test_final)

final_smape = final_sMAPE_calc(rouger_smape_test, final_smape_test)
print('–°–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è —Å—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ =', round(final_smape,3))
```
